```python
# Drop the unnamed index column
df=df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

```python
# Check how many rows have any missing values
print("Rows with missing values:", df.isnull().any(axis=1).sum())

# Drop rows with any missing values
df = df.dropna()
```

```python
df = df.drop_duplicates()
df
```

```python
# Convert text columns to lowercase and remove extra whitespace
df['Education'] = df['Education'].str.lower().str.strip()
df['Marital_Status'] = df['Marital_Status'].str.lower().str.strip()
print(df['education'],df['Marital_Status'] )
```

```python
# Convert date column to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce', format='%Y-%m-%d')
df["Dt_Customer"]
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
df['year_birth'] = df['year_birth'].astype(int)
df['kidhome'] = df['kidhome'].astype(int)
print(df['year_birth'],df['kidhome'])
```

```python
df.to_csv('customer_personality_cleaned.csv', index=False)
df
```
